```python
import os
import glob
import shutil

#new rename function which accounts for renamed files
def rename_my_file_new(old_file_name):
    month_dic = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    for value in month_dic.values():
        if old_file_name[:2] == value:
            return old_file_name
    original_text_month, original_text_rest = old_file_name.split('-')
    new_text_month = month_dic[original_text_month]
    new_text = new_text_month + '-' + original_text_rest.replace('_','-')
    return new_text

#Find all files that end with .txt
path_to_old_file = os.path.join('os-collaboration-exercise-data','*.txt')
list_of_old_file_paths = glob.glob(path_to_old_file, recursive=True)

#rename all the files in the list we found above
for old_file_path in list_of_old_file_paths:
    old_file_name = old_file_path.split('\\')[1] #obtain file name from path
    new_file_path = os.path.join('os-collaboration-exercise-data',rename_my_file_new(old_file_name)) #obtain new file path
    os.rename(old_file_path, new_file_path)
    print(f"{old_file_name} has been renamed to {rename_my_file_new(old_file_name)}!")
```

```python
#Find all files that end with cairo.txt
path_to_cairo_file = os.path.join('os-collaboration-exercise-data','*cairo.txt')
list_of_cairo_file_paths = glob.glob(path_to_cairo_file, recursive=True)

#move all the files in the list we found above
for old_cairo_file_path in list_of_cairo_file_paths:
    #find old file names
    old_cairo_file_name = old_cairo_file_path.split('\\')[-1] 
    #obtain new file path
    new_cairo_file_path = os.path.join('os-collaboration-exercise-data','Cairo',old_cairo_file_name) 
    shutil.copy(old_cairo_file_path, new_cairo_file_path)
    print(f"{old_cairo_file_name} has been copied to the folder {new_cairo_file_path.split('\\')[-2]}!")
```

## Task 7

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    #searching for city files
    city_file_name = '*' + city.lower() + '.txt' 
    path_to_city_file = os.path.join('os-collaboration-exercise-data',city_file_name) 
    list_of_city_file_paths = glob.glob(path_to_city_file, recursive=True)
    #move all the files in the list we found above
    for old_city_file_path in list_of_city_file_paths:
        old_city_file_name = old_city_file_path.split('\\')[-1]
        new_city_file_path = os.path.join('os-collaboration-exercise-data',city,old_city_file_name)
        shutil.move(old_city_file_path, new_city_file_path)
        print(f"{old_city_file_name} has been moved to {new_city_file_path.split('\\')[-2]}!")
```
