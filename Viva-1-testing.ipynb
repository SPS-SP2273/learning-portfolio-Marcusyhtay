```python
from matplotlib import pyplot as plt
import numpy as np
dna_seq = ''
for i in range (200):
    dna_seq += np.random.choice(['A','T','C','G'])

count = 0
a_pos = []
t_pos = []
c_pos = []
g_pos = []
for base in dna_seq:
    count += 1
    if base == 'A':
        a_pos.append(count)
    elif base == 'T':
        t_pos.append(count)
    elif base == 'C':
        c_pos.append(count)
    elif base == 'G':
        g_pos.append(count)
plt.figure(figsize=(15, 4)) 
plt.xlim(0,200)
plt.yticks([])
plt.hist([a_pos, t_pos,c_pos,g_pos], bins=200, label=['A', 'T','C','G'],histtype='stepfilled')
plt.xlabel('Nucleotide position (bp)')
plt.legend()
```

```python
from matplotlib import pyplot as plt
import numpy as np
dna_seq = ''
for i in range (200):
    dna_seq += np.random.choice(['A','T','C','G'])

count = 0
a_pos = []
t_pos = []
c_pos = []
g_pos = []
for base in dna_seq:
    count += 1
    if base == 'A':
        a_pos.append(count)
    elif base == 'T':
        t_pos.append(count)
    elif base == 'C':
        c_pos.append(count)
    elif base == 'G':
        g_pos.append(count)

plt.figure(figsize=(15, 2))
plt.scatter(a_pos, ['A']*len(a_pos), color='red', label='A')
plt.scatter(t_pos, ['T']*len(t_pos), color='blue', label='T')
plt.scatter(c_pos, ['C']*len(c_pos), color='green', label='C')
plt.scatter(g_pos, ['G']*len(g_pos), color='orange', label='G')
plt.xlim(0, 200)
plt.yticks(['A', 'T', 'C', 'G'])
plt.xlabel('Position in DNA sequence')
plt.legend()
plt.show()
```
